1. What is a RESTful API?
A RESTful API is a way for two computer systems to talk to each other over the internet using HTTP. It follows certain rules (REST) to create, read, update, or delete data.
2. Explain the concept of API specification
An API specification is a blueprint or guide that tells developers what endpoints are available, what data to send/receive, and which HTTP method to use.
3. What is Flask, and why is it popular for building APIs?
Flask is a Python web framework used to create web apps and APIs. It's popular because it's lightweight, simple, easy to learn, and flexible.
4. What is routing in Flask?
Routing matches URLs to specific Python functions. For example, @app.route('/hello') triggers a function when someone visits /hello.
5. How do you create a simple Flask application?
from flask import Flask
app = Flask(__name__)

@app.route('/')
def home():
    return 'Hello, Flask!'

if __name__ == '__main__':
    app.run()
6. What are HTTP methods used in RESTful APIs?
Common methods include GET (retrieve), POST (create), PUT (update), DELETE (remove).
7. What is the purpose of the @app.route() decorator in Flask?
It tells Flask which URL should trigger which function.
8. What is the difference between GET and POST HTTP methods?
GET fetches data, while POST sends new data (like form submissions).
9. How do you handle errors in Flask APIs?
Use try-except blocks or Flask’s error handlers like @app.errorhandler(404) to return custom messages.
10. How do you connect Flask to a SQL database?
Use Flask-SQLAlchemy. Example:
app.config['SQLALCHEMY_DATABASE_URI'] = 'sqlite:///example.db'
db = SQLAlchemy(app)
11. What is the role of Flask-SQLAlchemy?
It allows you to interact with SQL databases using Python classes instead of raw SQL queries.
12. What are Flask blueprints, and how are they useful?
Blueprints help you organize large apps into smaller, manageable parts like modules for users or admin.
13. What is the purpose of Flask's request object?
It allows access to data sent by the client, such as form inputs or JSON data.
14. How do you create a RESTful API endpoint using Flask?
@app.route('/api/user/<int:id>', methods=['GET'])
def get_user(id):
    return {'id': id, 'name': 'John'}
15. What is the purpose of Flask's jsonify() function?
It converts Python dictionaries into JSON format, which is commonly used in APIs.
16. Explain Flask’s url_for() function
url_for() dynamically generates URLs for routes, useful for linking and redirection.
17. How does Flask handle static files (CSS, JavaScript, etc.)?
Flask serves static files from the 'static/' folder. For example, /static/style.css.
18. What is an API specification, and how does it help in building a Flask API?
It defines how the API should work and helps developers know what endpoints exist and what data to send/receive.
19. What are HTTP status codes, and why are they important in a Flask API?
Status codes indicate the result of a request, like 200 (OK), 404 (Not Found), or 500 (Error).
20. How do you handle POST requests in Flask?
from flask import request
@app.route('/submit', methods=['POST'])
def submit():
    data = request.json
    return jsonify(data)
21. How would you secure a Flask API?
Use authentication (tokens), input validation, HTTPS, and rate limiting.
22. What is the significance of the Flask-RESTful extension?
Flask-RESTful helps you build REST APIs by organizing resources and providing helpful tools.
23. What is the role of Flask’s session object?
It stores user-specific data like login info across multiple requests using cookies.


1. How do you create a basic Flask application?
from flask import Flask
app = Flask(__name__)

@app.route('/')
def home():
    return 'Hello, World!'

if __name__ == '__main__':
    app.run(debug=True)
2. How do you serve static files like images or CSS in Flask?
Place files in a folder named 'static'. Access them using '/static/filename.ext'.
Example: <img src='/static/logo.png'>

3. How do you define different routes with different HTTP methods in Flask?
@app.route('/submit', methods=['GET', 'POST'])
def submit():
    if request.method == 'POST':
        return 'Form submitted'
    return 'Submit page'

4. How do you render HTML templates in Flask?
Use Flask's render_template function and put HTML files in a 'templates/' folder.
Example:
@app.route('/')
def home():
    return render_template('index.html')

5. How can you generate URLs for routes in Flask using url_for?
Use url_for('function_name') to generate the URL dynamically.
Example: url_for('home')

6. How do you handle forms in Flask?
Use the request object to get form data:
@app.route('/submit', methods=['POST'])
def submit():
    name = request.form['name']
    return f'Hello, {name}'

7. How can you validate form data in Flask?
Use simple checks or libraries like WTForms.
Example:
if not request.form['name']:
    return 'Name is required'

8. How do you manage sessions in Flask?
Use Flask’s session object:
session['user'] = 'John'
Access it using session.get('user')

9. How do you redirect to a different route in Flask?
Use redirect() with url_for():
return redirect(url_for('home'))

10. How do you handle errors in Flask (e.g., 404)?
@app.errorhandler(404)
def not_found(error):
    return 'Page not found', 404

11. How do you structure a Flask app using Blueprints?
Create a Blueprint in a separate file and register it:

from flask import Blueprint
my_bp = Blueprint('my_bp', __name__)

@my_bp.route('/')
def home():
    return 'Home from blueprint'

app.register_blueprint(my_bp, url_prefix='/bp')

12. How do you define a custom Jinja filter in Flask?
@app.template_filter('reverse')
def reverse_filter(s):
    return s[::-1]
Use in template: {{ 'hello'|reverse }}

13. How can you redirect with query parameters in Flask?
Use url_for with parameters:
redirect(url_for('search', q='flask'))

14. How do you return JSON responses in Flask?
Use jsonify():
from flask import jsonify
return jsonify({'name': 'John', 'age': 30})

15. How do you capture URL parameters in Flask?
@app.route('/user/<username>')
def show_user(username):
    return f'Hello {username}'
